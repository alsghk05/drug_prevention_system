```python
from sklearn.utils import resample

# SMOTE를 사용하여 각 열에 대해 데이터 증강
smote = SMOTE(sampling_strategy='auto', random_state=42)

# 결과를 저장할 데이터프레임 초기화
resampled_features = []
resampled_labels = []

# 각 마약 종류에 대해 SMOTE 적용
for column in df_labels.columns:
    features_res, labels_res = smote.fit_resample(df_features, df_labels[column])
    resampled_features.append(features_res)
    resampled_labels.append(labels_res)

# 가장 큰 샘플 수를 찾습니다.
max_samples = max([len(labels) for labels in resampled_labels])

# 각 레이블 컬럼의 크기를 맞춥니다.
resampled_features_balanced = resampled_features[0]  # 첫 번째 컬럼을 기준으로 합니다.
resampled_labels_balanced = []

for i, labels in enumerate(resampled_labels):
    if len(labels) < max_samples:
        features_res, labels_res = resample(
            resampled_features[i], labels,
            replace=True,
            n_samples=max_samples - len(labels),
            random_state=42
        )
        resampled_features_balanced = pd.concat([resampled_features_balanced, features_res], ignore_index=True)
        resampled_labels_balanced.append(pd.concat([pd.Series(labels), pd.Series(labels_res)], ignore_index=True))
    else:
        resampled_features_balanced = pd.concat([resampled_features_balanced, resampled_features[i]], ignore_index=True)
        resampled_labels_balanced.append(pd.Series(labels))

# 데이터 프레임 생성
df_features_resampled = pd.DataFrame(resampled_features_balanced, columns=df_features.columns)
df_labels_resampled = pd.concat(resampled_labels_balanced, axis=1)
df_labels_resampled.columns = df_labels.columns

# 증강된 데이터 확인
unique_value_counts_resampled = df_labels_resampled.apply(pd.Series.value_counts).fillna(0).astype(int)
unique_value_counts_resampled
```

### Neural Net 파라미터 튜닝

```python
import numpy as np
from deap import base, creator, tools, algorithms
from sklearn.multioutput import MultiOutputClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import KFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 데이터 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_features)

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X_scaled, df_modified_new, test_size=0.2, random_state=42)

# 평가 함수 정의
def evaluate(individual):
    # MLP 하이퍼파라미터 설정
    hidden_layer_sizes = (int(individual[0]), int(individual[1]))
    activation = ['identity', 'logistic', 'tanh', 'relu'][int(individual[2])]
    solver = ['lbfgs', 'sgd', 'adam'][int(individual[3])]
    alpha = individual[4]
    
    # 모델 정의
    model = MultiOutputClassifier(
        MLPClassifier(hidden_layer_sizes=hidden_layer_sizes, activation=activation, solver=solver, alpha=alpha, max_iter=1000, random_state=42)
    )
    
    # 교차 검증 수행
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    scores = []
    
    for train_index, val_index in kf.split(X_train):
        X_fold_train, X_fold_val = X_train[train_index], X_train[val_index]
        y_fold_train, y_fold_val = y_train.iloc[train_index], y_train.iloc[val_index]
        
        model.fit(X_fold_train, y_fold_train)
        y_fold_pred = model.predict(X_fold_val)
        
        fold_accuracy = np.mean([accuracy_score(y_fold_val.iloc[:, i], y_fold_pred[:, i]) for i in range(y_fold_val.shape[1])])
        scores.append(fold_accuracy)
    
    return np.mean(scores),

# DEAP 설정
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

toolbox = base.Toolbox()
toolbox.register("attr_int", np.random.randint, 50, 200)  # hidden_layer_sizes 범위
toolbox.register("attr_int_2", np.random.randint, 50, 200)  # hidden_layer_sizes 범위
toolbox.register("attr_int_3", np.random.randint, 0, 4)  # activation 함수 인덱스
toolbox.register("attr_int_4", np.random.randint, 0, 3)  # solver 인덱스
toolbox.register("attr_float", np.random.uniform, 0.0001, 1)  # alpha 값 범위
toolbox.register("individual", tools.initCycle, creator.Individual, 
                 (toolbox.attr_int, toolbox.attr_int_2, toolbox.attr_int_3, toolbox.attr_int_4, toolbox.attr_float), n=1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("mate", tools.cxUniform, indpb=0.5)
toolbox.register("mutate", tools.mutPolynomialBounded, low=[50, 50, 0, 0, 0.0001], up=[200, 200, 3, 2, 1], eta=0.1, indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("evaluate", evaluate)

# DEAP 메인 설정
def main():
    np.random.seed(42)
    
    # 인구 생성
    population = toolbox.population(n=50)
    
    # 알고리즘 설정
    NGEN = 10
    CXPB = 0.5
    MUTPB = 0.2
    
    # 진화 과정
    for gen in range(NGEN):
        offspring = algorithms.varAnd(population, toolbox, CXPB, MUTPB)
        fits = list(map(toolbox.evaluate, offspring))
        
        for fit, ind in zip(fits, offspring):
            ind.fitness.values = fit
        
        population = toolbox.select(offspring, k=len(population))
    
    # 최적 개체 선택
    best_ind = tools.selBest(population, 1)[0]
    return best_ind

# 적합된 모델 도출
best_params = main()
print(f"Best Parameters: hidden_layer_sizes=({int(best_params[0])}, {int(best_params[1])}), activation={['identity', 'logistic', 'tanh', 'relu'][int(best_params[2])]}, solver={['lbfgs', 'sgd', 'adam'][int(best_params[3])]}, alpha={best_params[4]}")

# 최적 하이퍼파라미터를 사용한 모델 학습 및 평가
best_model = MultiOutputClassifier(
    MLPClassifier(hidden_layer_sizes=(int(best_params[0]), int(best_params[1])), 
                  activation=['identity', 'logistic', 'tanh', 'relu'][int(best_params[2])], 
                  solver=['lbfgs', 'sgd', 'adam'][int(best_params[3])], 
                  alpha=best_params[4], max_iter=1000, random_state=42)
)
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)

# 성능 지표 계산
test_accuracy = np.mean([accuracy_score(y_test.iloc[:, i], y_pred[:, i]) for i in range(y_test.shape[1])])
test_precision = np.mean([precision_score(y_test.iloc[:, i], y_pred[:, i], average='weighted') for i in range(y_test.shape[1])])
test_recall = np.mean([recall_score(y_test.iloc[:, i], y_pred[:, i], average='weighted') for i in range(y_test.shape[1])])
test_f1_score = np.mean([f1_score(y_test.iloc[:, i], y_pred[:, i], average='weighted') for i in range(y_test.shape[1])])

print(f"Test Set Accuracy: {test_accuracy:.4f}")
print(f"Test Set Precision: {test_precision:.4f}")
print(f"Test Set Recall: {test_recall:.4f}")
print(f"Test Set F1 Score: {test_f1_score:.4f}")
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
from sklearn.neural_network import MLPClassifier

## 지금 최종최종최종
mlp_clf_tuned = MLPClassifier(
    activation = 'relu',
    solver = 'adam',
    hidden_layer_sizes = (50,),
    alpha = 0.06357208758199738,
    random_state=42
)

# MultiOutputClassifier로 MLPClassifier 래핑
multi_model = MultiOutputClassifier(mlp_clf_tuned)

# 모델 학습
multi_model.fit(X_train, y_train)

# 예측 수행
y_pred_multi_mlp = multi_model.predict(X_test)

# 결과 저장을 위한 딕셔너리 초기화
multioutput_results_mlp_tuned = {}

# Numpy 배열로 변환
y_test_np = y_test.values
y_pred_multi_mlp_np = y_pred_multi_mlp

# 각 레이블에 대한 지표 계산 후 평균 취하기
accuracy_avg_mlp_tuned = np.mean([accuracy_score(y_test_np[:, i], y_pred_multi_mlp_np[:, i])
                                  for i in range(y_test_np.shape[1])])
f1_avg_mlp_tuned = np.mean([f1_score(y_test_np[:, i], y_pred_multi_mlp_np[:, i], average='weighted')
                            for i in range(y_test_np.shape[1])])
recall_avg_mlp_tuned = np.mean([recall_score(y_test_np[:, i], y_pred_multi_mlp_np[:, i], average='weighted')
                                for i in range(y_test_np.shape[1])])
precision_avg_mlp_tuned = np.mean([precision_score(y_test_np[:, i], y_pred_multi_mlp_np[:, i], average='weighted')
                                   for i in range(y_test_np.shape[1])])

# 결과 저장
multioutput_results_mlp_tuned['MLPClassifier'] = {
    'Accuracy': accuracy_avg_mlp_tuned,
    'F1 Score': f1_avg_mlp_tuned,
    'Recall': recall_avg_mlp_tuned,
    'Precision': precision_avg_mlp_tuned
}

print("Tuned MLPClassifier Multi-output Classification Results:")
print(multioutput_results_mlp_tuned)
```
