## 신경망 하이퍼파라미터 튜닝에 사용한 것들
- activation
- optimizer
- learning rate of optimizer (그래프 필수)
- dropout rate (그래프 필수)
- epoch and batch_size
- hidden layer and neurons (제외해도 됨)

**learning rate랑 dropout rate 조절을 위주로 시각화 자료 들어가고 상세 설명 들어갈 예정**
+ 기타 모델들은 비교 대상을 들어갈 것
+ 비교한 결과는 Accuracy, F1 score, Recall, Precision, (AUC) 등등 평가 지표를 표로 나타내야 함!
+ 찬기님 논문 참고하면 표 있어용
