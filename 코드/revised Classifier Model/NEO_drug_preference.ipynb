```python
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.neural_network import MLPClassifier
import numpy as np
```

```python
# 데이터 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_features)
```

```python
# 분류기 정의
classifiers = {
    "Logistic Regression": MultiOutputClassifier(LogisticRegression(max_iter=1000)),
    "K-Nearest Neighbors": MultiOutputClassifier(KNeighborsClassifier()),
    "Support Vector Machine": MultiOutputClassifier(SVC()),
    "Random Forest": MultiOutputClassifier(RandomForestClassifier()),
    "Gradient Boosting": MultiOutputClassifier(GradientBoostingClassifier()),
    "Neural Network": MultiOutputClassifier(MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42))
}

# 10-폴드 교차 검증 설정
skf = StratifiedKFold(n_splits=10)
```

```python
# 각 분류기에 대해 성능 비교
results = {
    "Model": [],
    "Fold": [],
    "Accuracy": [],
    "Precision": [],
    "Recall": [],
    "F1 Score": []
}

for name, clf in classifiers.items():
    fold_idx = 1
    for train_index, test_index in skf.split(X_scaled, df_modified_new.values.argmax(axis=1)):
        X_train_fold, X_test_fold = X_scaled[train_index], X_scaled[test_index]
        y_train_fold, y_test_fold = df_modified_new.values[train_index], df_modified_new.values[test_index]
        
        clf.fit(X_train_fold, y_train_fold)
        y_pred_fold = clf.predict(X_test_fold)
        
        accuracy = np.mean([accuracy_score(y_test_fold[:, i], y_pred_fold[:, i]) for i in range(y_test_fold.shape[1])])
        precision = np.mean([precision_score(y_test_fold[:, i], y_pred_fold[:, i], average='weighted') for i in range(y_test_fold.shape[1])])
        recall = np.mean([recall_score(y_test_fold[:, i], y_pred_fold[:, i], average='weighted') for i in range(y_test_fold.shape[1])])
        f1 = np.mean([f1_score(y_test_fold[:, i], y_pred_fold[:, i], average='weighted') for i in range(y_test_fold.shape[1])])
        
        results["Model"].append(name)
        results["Fold"].append(fold_idx)
        results["Accuracy"].append(accuracy)
        results["Precision"].append(precision)
        results["Recall"].append(recall)
        results["F1 Score"].append(f1)
        
        fold_idx += 1

# 데이터 프레임으로 변환
results_df = pd.DataFrame(results)

# 각 모델에 대해 10-폴드 결과의 평균 계산
average_results = results_df.groupby('Model').mean().reset_index()

# 결과 출력
print(average_results)

# 최종 모델 선택 및 전체 데이터셋에 대해 성능 평가
best_classifier_name = average_results.groupby('Model')['F1 Score'].mean().idxmax()
best_classifier = classifiers[best_classifier_name]
best_classifier.fit(X_scaled, df_modified_new)
y_pred = best_classifier.predict(X_scaled)
test_accuracy = np.mean([accuracy_score(df_modified_new.iloc[:, i], y_pred[:, i]) for i in range(df_modified_new.shape[1])])
test_precision = np.mean([precision_score(df_modified_new.iloc[:, i], y_pred[:, i], average='weighted') for i in range(df_modified_new.shape[1])])
test_recall = np.mean([recall_score(df_modified_new.iloc[:, i], y_pred[:, i], average='weighted') for i in range(df_modified_new.shape[1])])
test_f1_score = np.mean([f1_score(df_modified_new.iloc[:, i], y_pred[:, i], average='weighted') for i in range(df_modified_new.shape[1])])

print(f"Best Classifier: {best_classifier_name}")
print(f"Test Set Accuracy: {test_accuracy:.4f}")
print(f"Test Set Precision: {test_precision:.4f}")
print(f"Test Set Recall: {test_recall:.4f}")
print(f"Test Set F1 Score: {test_f1_score:.4f}")

# 평균 결과 데이터 프레임 반환
average_results
```

```python
from tpot import TPOTClassifier
import optuna

# 최적화 결과를 저장할 데이터 프레임
optimized_params = []
optimized_scores = []

# 분류기 정의 및 Optuna 최적화
def optimize_logistic_regression(trial):
    penalty = trial.suggest_categorical('penalty', ['l1', 'l2', 'elasticnet', 'none'])
    C = trial.suggest_float('C', 1e-4, 1e2, log=True)
    solver = trial.suggest_categorical('solver', ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'])
    clf = LogisticRegression(penalty=penalty, C=C, solver=solver, max_iter=1000)
    score = cross_val_score(clf, X_scaled, df_modified_new.values.argmax(axis=1), cv=5, scoring='f1_weighted').mean()
    return score

def optimize_knn(trial):
    n_neighbors = trial.suggest_int('n_neighbors', 1, 50)
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    score = cross_val_score(clf, X_scaled, df_modified_new.values.argmax(axis=1), cv=5, scoring='f1_weighted').mean()
    return score

def optimize_svc(trial):
    C = trial.suggest_float('C', 1e-4, 1e2, log=True)
    kernel = trial.suggest_categorical('kernel', ['linear', 'poly', 'rbf', 'sigmoid'])
    clf = SVC(C=C, kernel=kernel)
    score = cross_val_score(clf, X_scaled, df_modified_new.values.argmax(axis=1), cv=5, scoring='f1_weighted').mean()
    return score

def optimize_random_forest(trial):
    n_estimators = trial.suggest_int('n_estimators', 10, 1000)
    max_depth = trial.suggest_int('max_depth', 1, 30)
    clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth)
    score = cross_val_score(clf, X_scaled, df_modified_new.values.argmax(axis=1), cv=5, scoring='f1_weighted').mean()
    return score

def optimize_gradient_boosting(trial):
    n_estimators = trial.suggest_int('n_estimators', 10, 1000)
    learning_rate = trial.suggest_float('learning_rate', 1e-4, 1.0, log=True)
    clf = GradientBoostingClassifier(n_estimators=n_estimators, learning_rate=learning_rate)
    score = cross_val_score(clf, X_scaled, df_modified_new.values.argmax(axis=1), cv=5, scoring='f1_weighted').mean()
    return score

def optimize_mlp(trial):
    hidden_layer_sizes = trial.suggest_categorical('hidden_layer_sizes', [(50,), (100,), (50, 50), (100, 50)])
    activation = trial.suggest_categorical('activation', ['identity', 'logistic', 'tanh', 'relu'])
    solver = trial.suggest_categorical('solver', ['lbfgs', 'sgd', 'adam'])
    alpha = trial.suggest_float('alpha', 1e-5, 1e-1, log=True)
    clf = MLPClassifier(hidden_layer_sizes=hidden_layer_sizes, activation=activation, solver=solver, alpha=alpha, max_iter=1000)
    score = cross_val_score(clf, X_scaled, df_modified_new.values.argmax(axis=1), cv=5, scoring='f1_weighted').mean()
    return score

# 각 분류기에 대해 최적화 수행
studies = {
    "Logistic Regression": optimize_logistic_regression,
    "K-Nearest Neighbors": optimize_knn,
    "Support Vector Machine": optimize_svc,
    "Random Forest": optimize_random_forest,
    "Gradient Boosting": optimize_gradient_boosting,
    "Neural Network": optimize_mlp
}

for name, optimize_func in studies.items():
    print(f"Optimizing {name}...")
    study = optuna.create_study(direction='maximize')
    study.optimize(optimize_func, n_trials=50)
    optimized_params.append({'Model': name, 'Best Params': study.best_params})
    clf = study.best_trial.user_attrs['clf']
    clf.fit(X_scaled, df_modified_new.values.argmax(axis=1))
    y_pred = clf.predict(X_scaled)
    optimized_scores.append({
        'Model': name,
        'Accuracy': accuracy_score(df_modified_new.values.argmax(axis=1), y_pred),
        'Precision': precision_score(df_modified_new.values.argmax(axis=1), y_pred, average='weighted'),
        'Recall': recall_score(df_modified_new.values.argmax(axis=1), y_pred, average='weighted'),
        'F1 Score': f1_score(df_modified_new.values.argmax(axis=1), y_pred, average='weighted')
    })

# 최적의 하이퍼 파라미터와 평가지표 결과를 데이터 프레임으로 변환
optimized_params_df = pd.DataFrame(optimized_params)
optimized_scores_df = pd.DataFrame(optimized_scores)

# 결과 출력
print("Optimized Parameters:")
print(optimized_params_df)
print("\nOptimized Scores:")
print(optimized_scores_df)

# 결과 데이터 프레임 반환
optimized_params_df, optimized_scores_df
```

### Neural Net 파라미터 튜닝

```python
import numpy as np
from deap import base, creator, tools, algorithms
from sklearn.multioutput import MultiOutputClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import KFold, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 데이터 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_features)

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X_scaled, df_modified_new, test_size=0.2, random_state=42)

# 평가 함수 정의
def evaluate(individual):
    # MLP 하이퍼파라미터 설정
    hidden_layer_sizes = (int(individual[0]), int(individual[1]))
    activation = ['identity', 'logistic', 'tanh', 'relu'][int(individual[2])]
    solver = ['lbfgs', 'sgd', 'adam'][int(individual[3])]
    alpha = individual[4]
    
    # 모델 정의
    model = MultiOutputClassifier(
        MLPClassifier(hidden_layer_sizes=hidden_layer_sizes, activation=activation, solver=solver, alpha=alpha, max_iter=1000, random_state=42)
    )
    
    # 교차 검증 수행
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    scores = []
    
    for train_index, val_index in kf.split(X_train):
        X_fold_train, X_fold_val = X_train[train_index], X_train[val_index]
        y_fold_train, y_fold_val = y_train.iloc[train_index], y_train.iloc[val_index]
        
        model.fit(X_fold_train, y_fold_train)
        y_fold_pred = model.predict(X_fold_val)
        
        fold_accuracy = np.mean([accuracy_score(y_fold_val.iloc[:, i], y_fold_pred[:, i]) for i in range(y_fold_val.shape[1])])
        scores.append(fold_accuracy)
    
    return np.mean(scores),

# DEAP 설정
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

toolbox = base.Toolbox()
toolbox.register("attr_int", np.random.randint, 50, 200)  # hidden_layer_sizes 범위
toolbox.register("attr_int_2", np.random.randint, 50, 200)  # hidden_layer_sizes 범위
toolbox.register("attr_int_3", np.random.randint, 0, 4)  # activation 함수 인덱스
toolbox.register("attr_int_4", np.random.randint, 0, 3)  # solver 인덱스
toolbox.register("attr_float", np.random.uniform, 0.0001, 1)  # alpha 값 범위
toolbox.register("individual", tools.initCycle, creator.Individual, 
                 (toolbox.attr_int, toolbox.attr_int_2, toolbox.attr_int_3, toolbox.attr_int_4, toolbox.attr_float), n=1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("mate", tools.cxUniform, indpb=0.5)
toolbox.register("mutate", tools.mutPolynomialBounded, low=[50, 50, 0, 0, 0.0001], up=[200, 200, 3, 2, 1], eta=0.1, indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("evaluate", evaluate)

# DEAP 메인 설정
def main():
    np.random.seed(42)
    
    # 인구 생성
    population = toolbox.population(n=50)
    
    # 알고리즘 설정
    NGEN = 10
    CXPB = 0.5
    MUTPB = 0.2
    
    # 진화 과정
    for gen in range(NGEN):
        offspring = algorithms.varAnd(population, toolbox, CXPB, MUTPB)
        fits = list(map(toolbox.evaluate, offspring))
        
        for fit, ind in zip(fits, offspring):
            ind.fitness.values = fit
        
        population = toolbox.select(offspring, k=len(population))
    
    # 최적 개체 선택
    best_ind = tools.selBest(population, 1)[0]
    return best_ind

# 적합된 모델 도출
best_params = main()
print(f"Best Parameters: hidden_layer_sizes=({int(best_params[0])}, {int(best_params[1])}), activation={['identity', 'logistic', 'tanh', 'relu'][int(best_params[2])]}, solver={['lbfgs', 'sgd', 'adam'][int(best_params[3])]}, alpha={best_params[4]}")

# 최적 하이퍼파라미터를 사용한 모델 학습 및 평가
best_model = MultiOutputClassifier(
    MLPClassifier(hidden_layer_sizes=(int(best_params[0]), int(best_params[1])), 
                  activation=['identity', 'logistic', 'tanh', 'relu'][int(best_params[2])], 
                  solver=['lbfgs', 'sgd', 'adam'][int(best_params[3])], 
                  alpha=best_params[4], max_iter=1000, random_state=42)
)
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)

# 성능 지표 계산
test_accuracy = np.mean([accuracy_score(y_test.iloc[:, i], y_pred[:, i]) for i in range(y_test.shape[1])])
test_precision = np.mean([precision_score(y_test.iloc[:, i], y_pred[:, i], average='weighted') for i in range(y_test.shape[1])])
test_recall = np.mean([recall_score(y_test.iloc[:, i], y_pred[:, i], average='weighted') for i in range(y_test.shape[1])])
test_f1_score = np.mean([f1_score(y_test.iloc[:, i], y_pred[:, i], average='weighted') for i in range(y_test.shape[1])])

print(f"Test Set Accuracy: {test_accuracy:.4f}")
print(f"Test Set Precision: {test_precision:.4f}")
print(f"Test Set Recall: {test_recall:.4f}")
print(f"Test Set F1 Score: {test_f1_score:.4f}")
```
