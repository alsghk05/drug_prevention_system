```python
import pandas as pd

df = pd.read_csv('drug_consumption.csv')
df = df.replace('CL0', 0).replace('CL1', 1).replace('CL2', 2).replace('CL3', 3).replace('CL4', 4).replace('CL5', 5).replace('CL6', 6)
score = df[['Amphet', 'Amyl', 'Benzos', 'Cannabis', 'Coke', 'Crack',
            'Ecstasy', 'Heroin', 'Ketamine', 'Nicotine', 'Legalh', 'LSD', 'Meth',
            'Semer', 'VSA']].sum(axis = 1)
df['score'] = score
# score 열의 값들을 퍼센트로 변환
# 각 점수를 백분율로 변환
df['score'] = (df['score'] / 90) * 100
df
```

```python
df = df.drop(['ID', 'Gender','Ethnicity', 'Alcohol',
       'Amphet', 'Amyl', 'Benzos', 'Caff', 'Cannabis', 'Choc', 'Coke', 'Crack',
       'Ecstasy', 'Heroin', 'Ketamine', 'Legalh', 'LSD', 'Meth', 'Mushrooms',
       'Nicotine', 'Semer', 'VSA'], axis = 1)
df
```

```python
dataset = df.values
print(dataset.shape)
```

```python
# 필요한 열 선택
X = dataset[:, 0:10]  # Age, Education, NeoScore, Implusive, SS
y = dataset[:, 10]    # score
```

```python
# 정규화
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# 분배
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)
```

# 신경망 모델

```python
import tensorflow as tf
import random
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Activation

np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

model = Sequential()
model.add(Dense(16, input_dim=10))
model.add(BatchNormalization())
model.add(Activation('selu'))
model.add(Dropout(0.0326530612244898))
model.add(Dense(16))
model.add(BatchNormalization())
model.add(Activation('selu'))
model.add(Dropout(0.0326530612244898))
model.add(Dense(1, activation='linear'))

# Compile the model
optimizer = RMSprop(learning_rate=0.004177551020408164)
model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mean_squared_error', 'mean_absolute_error'])

# Train the model
history = model.fit(X_train, y_train, epochs=24, batch_size=40, validation_data=(X_test, y_test), verbose=0)

# Evaluate the model
loss, mse, mae = model.evaluate(X_test, y_test, verbose=0)

# RMSE
rmse = np.sqrt(mse)

# Append results
loss, mse, mae, rmse
```
