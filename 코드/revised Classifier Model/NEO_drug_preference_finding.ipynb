## 데이터 전처리

1) 불필요한 열 제거

```python
df.columns
```

```python
df = df.drop(['Unnamed: 0', 'ID', 'Age', 'Country', 'Education', 'Ethnicity', 'SS'], axis=1)
df.head()
```

2. 데이터 스케일링

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 수치형 데이터 선택
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_data = df[numeric_columns]

# Standard Scaler 초기화 및 적용
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# 스케일링된 데이터를 DataFrame으로 변환
scaled_data_df = pd.DataFrame(scaled_data, columns=numeric_columns)

# 결과 확인
scaled_data_df.head()
```

3. 범주형 데이터 인코딩

```python
df = df.replace('CL0', 0).replace('CL1', 1).replace('CL2', 2).replace('CL3', 3).replace('CL4', 4).replace('CL5', 5).replace('CL6', 6)
df
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 수치형 데이터 선택
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_data = df[numeric_columns]

# Standard Scaler 초기화 및 적용
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# 스케일링된 데이터를 DataFrame으로 변환
scaled_data_df2 = pd.DataFrame(scaled_data, columns=numeric_columns)

# 결과 확인
scaled_data_df2.head()
```

- df_features에 성격 검사 점수를 넣고, df_labels에 각 마약 종류의 소비 빈도를 넣음

```python
df_features = df[['Gender', 'Escore', 'Oscore', 'Ascore', 'Cscore', 'Nscore', 'Impulsive']]
df_labels = df[['Amphet', 'Benzos', 'Cannabis', 'Coke', 'Crack', 'Ecstasy', 'Heroin', 'Ketamine',
        'LSD', 'Meth', 'VSA', 'Mushrooms']]
df_labels_updown = df[['Amphet', 'Benzos', 'Cannabis', 'Coke', 'Crack', 'Ecstasy', 'Heroin', 'Ketamine',
        'LSD', 'Meth', 'VSA', 'Mushrooms']]
```

- 마약 종류 3가지로 분류

```python
# 마약 종류를 분류
stimulants = ['Amphet', 'Coke', 'Crack']
depressants = ['Benzos', 'Heroin', 'Ketamine', 'Meth','VSA']
psychedelics = ['Cannabis', 'Ecstasy', 'LSD', 'Mushrooms']

# 열 이름을 4가지 마약 종류로 변경
df_labels_updown.columns = ['Stimulants' if col in stimulants else 
                     'Depressants' if col in depressants else 
                     'Psychedelics' if col in psychedelics else col for col in df_labels.columns]
```
